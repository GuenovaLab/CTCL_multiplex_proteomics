```python
import os
import pandas as pd
import numpy as np
from tifffile import imread
import random
import re

%gui qt
```

```python
import napari
from napari.utils import nbscreenshot

# Create an empty viewer
viewer = napari.Viewer()
```

```python
# Change below
output_dir = "../output/"
tiff_dir = "/mnt/RECHERCHE/GUENOVA_LAB/Project_Multiplex_Phenotyping/Miltenyi/20230601/"
marker_file  = "..//annotation/marker_metadata.csv"
marker_df = pd.read_csv(os.path.join(marker_file))
marker_unique = marker_df.Marker[(marker_df.Marker != "DAPI") & (marker_df.PassOverallQuality == True)]
marker_unique = marker_unique.values
contrast_min = marker_df.ContrastRange_min_20230601[(marker_df.Marker != "DAPI") & (marker_df.PassOverallQuality == True)]
contrast_min = contrast_min.values
contrast_max = marker_df.ContrastRange_max_20230601[(marker_df.Marker != "DAPI") & (marker_df.PassOverallQuality == True)]
contrast_max = contrast_max.values
```

```python
marker_unique
```

```python
# CHANGE HERE SAMPLE :
samples = []

for i in range(28,63):
    samples = samples + ["ROI-" + str(i + 1)]
    
range_marker = 0
```

```python
range_marker
```

```python
@viewer.bind_key('Ctrl-Shift-D', overwrite = "True")
def visible_1(viewer):
    global marker_unique
    global range_marker
        
    print(marker_unique[range_marker])
    for sample in samples:
        print(sample)
        if (marker_unique[range_marker] == "TRBC1"):
            regex = re.compile("-" + marker_unique[range_marker])
        else:
            regex = re.compile("-" + marker_unique[range_marker] + "_")
        for file in os.listdir(os.path.join(tiff_dir, sample)):
            if regex.search(file) is not None:
                image = imread(os.path.join(tiff_dir, sample, file))
                viewer.add_image(image, name=sample + "-" + marker_unique[range_marker],
                                 contrast_limits = [contrast_min[range_marker], contrast_max[range_marker]]);
    range_marker = range_marker + 1
```

```python
@viewer.bind_key('Ctrl-Shift-L', overwrite = "True")
def visible_1(viewer):
    global marker_unique
    global range_marker
    
    
    layers = viewer.layers
    for layer in layers:
        viewer.layers.remove(value = layer.name)
        
    layers = viewer.layers
    for layer in layers:
        viewer.layers.remove(value = layer.name)
        
    layers = viewer.layers
    for layer in layers:
        viewer.layers.remove(value = layer.name)
        
    print(marker_unique[range_marker])
    for sample in samples:
        print(sample)
        regex = re.compile("-" + "DAPI")
        for file in os.listdir(os.path.join(tiff_dir, sample)):
            if regex.search(file) is not None:
                image = imread(os.path.join(tiff_dir, sample, file))
                viewer.add_image(image, name=sample + "-" + "DAPI", colormap="bop blue", contrast_limits=[5000, 65000]);

        if (marker_unique[range_marker] == "TRBC1"):
            regex = re.compile("-" + marker_unique[range_marker])
        else:
            regex = re.compile("-" + marker_unique[range_marker] + "_")
        for file in os.listdir(os.path.join(tiff_dir, sample)):
            if regex.search(file) is not None:
                image = imread(os.path.join(tiff_dir, sample, file))
                viewer.add_image(image, name=sample + "-" + marker_unique[range_marker], blending = "additive",
                                 contrast_limits = [contrast_min[range_marker], contrast_max[range_marker]]);
    range_marker = range_marker + 1
```
