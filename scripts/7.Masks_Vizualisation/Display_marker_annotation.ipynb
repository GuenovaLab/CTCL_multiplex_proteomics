```python
from skimage.filters import (threshold_otsu, threshold_niblack,
                             threshold_sauvola)
binary_global = image > threshold_otsu(image)

window_size = 49
thresh_niblack = threshold_niblack(image, window_size=window_size, k=0.01)
thresh_sauvola = threshold_sauvola(image, window_size=window_size)

binary_niblack = image > thresh_niblack
binary_sauvola = image > thresh_sauvola

viewer.add_image(binary_global, name=sample + "-" + marker + "-otsu")
viewer.add_image(binary_niblack, name=sample + "-" + marker + "-binary_niblack")
viewer.add_image(binary_sauvola, name=sample + "-" + marker + "-binary_sauvola")
```
